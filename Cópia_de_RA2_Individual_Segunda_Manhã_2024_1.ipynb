Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

**Atenção**: O tempo é importante, então escolha, entre usar ferramentas computacionais ou gráficas. Independente da sua escolha você deverá escrever a resposta em uma célula de textos.


# Resposta
a) $f'(x)=2 e^{2 x} - e^x$

$f'(x)=0$</br>
$0 = 2 e^{2 x} - e^x$</br>
$x=0$


```python
import sympy as sp
x, e = sp.symbols('x e')

f_x = sp.exp(2*x) - sp.exp(x)

# primeira derivada
f_1 = sp.diff(f_x, x)

#encontrando as raízes
critical_points = sp.solve(f_1, x)

# calculando a segunda derivada
f_2 = sp.diff(f_1, x)

for point in critical_points:
    second_derivative_value = f_2.subs(x, point)
    if second_derivative_value > 0:
        print(f"O ponto crítico em x = {point} é um mínimo local.")
    elif second_derivative_value < 0:
        print(f"O ponto crítico em x = {point} é um máximo local.")
```
