## Problema 1

$$\lim_{x \to 3} \frac{(x^2 - 9)}{(x - 3)}$$

Para resolver o limite

$$\lim_{x \to 3} \frac{x^2 - 9}{x - 3}$$

primeiro observamos que, ao substituir diretamente $x = 3$, obtemos uma forma indeterminada $\frac{0}{0}$.

Para simplificar podemos observar que o numerador é uma diferença de quadrados, que pode ser fatorada como:

$$x^2 - 9 = (x + 3)(x - 3)$$

Assim, a expressão se torna:

$$\frac{(x + 3)(x - 3)}{x - 3}$$

Observamos que o termo $(x - 3)$ no numerador e no denominador se cancelam, desde que $x \neq 3$ para evitar a divisão por zero. Portanto, a expressão simplificada é:

$$x + 3$$

Agora, podemos aplicar o limite:

$$\lim_{x \to 3} (x + 3) = 3 + 3 = 6$$

Portanto, o limite quando $x$ tende a 3 é 6.


Gráfico

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Define a função
def f(x):
    return (x**2 - 9) / (x - 3)

# Cria a variável simbólica x e define a função simbólica
x_sym = sp.symbols('x')
funcao_sym = (x_sym**2 - 9) / (x_sym - 3)


ponto = 3
x_start = -2
x_stop = 4
# Calcula o limite da função no ponto desejado
limite = sp.limit(funcao_sym, x_sym, ponto)

# Gera valores de x próximos a 3, evitando o valor exato 3 para evitar divisão por zero
x_vals = np.linspace(x_start, x_stop, 400)
x_vals = x_vals[x_vals != 3]

# Calcula os valores de y para os valores de x
y_vals = f(x_vals)

# Plota a função
plt.plot(x_vals, y_vals, label=r'$f(x) = \frac{x^2 - 9}{x - 3}$')

# Adiciona um círculo sem preenchimento na coordenada (3, limite) para indicar a descontinuidade
plt.scatter([3], [limite], s=30, facecolors='white', edgecolors='black')

# Linha vertical limitada ao valor do limite
plt.plot([ponto, ponto], [0, limite], 'orange', linestyle='--')

# Linha horizontal limitada ao ponto desejado
plt.plot([0, ponto], [limite, limite], 'orange', linestyle='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Gráfico de $f(x) = \frac{x^2 - 9}{x - 3}$', loc='right')
plt.legend()

# Adiciona seta no eixo x
plt.annotate('', xy=(max(x_vals), 0), xytext=(min(x_vals), 0),
             arrowprops=dict(arrowstyle="->", color='green'))

# Adiciona seta no eixo y
plt.annotate('', xy=(0, max(y_vals)+1), xytext=(0, min(y_vals)-2),
             arrowprops=dict(arrowstyle="->", color='green'))

# Coloca as letras x e y
plt.text(max(x_vals), 0, 'x', ha='left', va='center', fontsize=12)
plt.text(0, max(y_vals)+1, 'y', ha='center', va='bottom', fontsize=12)

# Adiciona uma label no ponto (3, limite)
plt.text(3.1, limite - 0.3, f'({3}, {limite})', fontsize=10)

# Ajusta os limites dos eixos para melhor visualização
plt.xlim(x_start, x_stop)
plt.ylim(min(y_vals)-2, max(y_vals)+1)

# Define os ticks para os eixos x e y
plt.xticks(np.arange(x_start, x_stop, 0.5))
plt.yticks(np.arange(min(y_vals)-2, max(y_vals)+1, 0.5))

# Ativa a grade
plt.grid(True, which='both', axis='both', color='lightgray', linestyle='-', linewidth=0.5)

plt.show()
```

## Problema 2

$$\lim_{x \to 2} \frac{x^3 - 8}{x^2 - 4}$$

Começamos com a substituição direta de $x = 2$ na equação:

$$\frac{2^3 - 8}{2^2 - 4} = \frac{0}{0}$$

Chegamos a uma indeterminação do tipo $\frac{0}{0}$. Para resolver essa indeterminação, vamos simplificar a expressão.

Podemos fatorar tanto o numerador quanto o denominador:

- O numerador $x^3 - 8$ pode ser fatorado como $(x - 2)(x^2 + 2x + 4)$, usando a fatoração de um cubo.
- O denominador $x^2 - 4$ pode ser fatorado como $(x - 2)(x + 2)$, usando a fatoração de uma diferença de quadrados.

Assim, a expressão pode ser simplificada para:

$$\lim_{x \to 2} \frac{(x - 2)(x^2 + 2x + 4)}{(x - 2)(x + 2)}$$

Cancelando o termo $(x - 2)$ comum ao numerador e ao denominador:

$$\lim_{x \to 2} \frac{x^2 + 2x + 4}{x + 2}$$

Agora, aplicamos a substituição direta novamente:

$$\frac{2^2 + 2 \cdot 2 + 4}{2 + 2} = \frac{4 + 4 + 4}{4} = \frac{12}{4} = 3$$

Portanto, o limite quando $x$ tende a 2 é 3.


## Problema 3

$$\lim_{x \to -1} \frac{x^3 + 2x^2 - x - 2}{x + 1}$$

Começamos com a substituição direta de $x = -1$ na expressão para ver se o limite é definido:

$$\lim_{x \to -1} \frac{x^3 + 2x^2 - x - 2}{x + 1} = \frac{(-1)^3 + 2(-1)^2 - (-1) - 2}{-1 + 1}$$

Isso nos dá:

$$\frac{-1 + 2 + 1 - 2}{0} = \frac{0}{0}$$

E chegamos a forma indeterminada $\frac{0}{0}$, precisamos simplificar a expressão para encontrar o limite.

Para remover a indefinição, vamos fatorar o numerador:

A expressão original do numerador é $x^3 + 2x^2 - x - 2$. Podemos tentar fatorá-la por agrupamento ou por divisão sintética, sabendo que $x + 1$ deve ser um fator, já que causa a indefinição por estar no denominador.

Fatorando o numerador:

$$x^3 + 2x^2 - x - 2 = (x + 1)(x^2 + x - 2)$$

Onde $x^2 + x - 2$ pode ser ainda mais fatorado:

$$x^2 + x - 2 = (x + 2)(x - 1)$$

Portanto, a expressão original pode ser reescrita como:

$$\lim_{x \to -1} \frac{(x + 1)(x + 2)(x - 1)}{x + 1}$$

Cancelamos $(x + 1)$ no numerador e no denominador:

$$\lim_{x \to -1} (x + 2)(x - 1)$$

Agora, voltamos a substituição direta e substituímos $x = -1$ na expressão simplificada:

$$\lim_{x \to -1} (-1 + 2)(-1 - 1) = (1)(-2) = -2$$

Portanto, o limite é $-2$.


## Problema 4

$$\lim_{x \to -4} \frac{x^2 - 5x + 4}{x^2 + 3x - 4}$$

Podemos fatorar tanto o numerador quanto o denominador.

O numerador $x^2 - 5x + 4$ fatora para $(x - 4)(x - 1)$, e o denominador $x^2 + 3x - 4$ fatora para $(x + 4)(x - 1)$. Portanto, a expressão original pode ser reescrita como:

$$\frac{(x - 4)(x - 1)}{(x + 4)(x - 1)}$$

Nesta forma, podemos ver que tanto o numerador quanto o denominador compartilham um fator comum $(x - 1)$. Cuidado ao cancelar este termo, pois estamos interessados no comportamento da função à medida que $x$ se aproxima de $-4$, e não simplesmente em simplificar a expressão.

A presença de $(x + 4)$ no denominador é o que causa uma problema que precisamos observar quando $x = -4$, já que isso resultaria em uma divisão por zero.

Ao avaliar o limite sem cancelar precipitadamente os termos comuns, notamos que o verdadeiro ponto de interesse é a indeterminação causada pelo termo $(x + 4)$ no denominador.

Quando $x$ se aproxima de $-4$, o denominador tende a zero, enquanto o numerador se torna um valor finito ($(-4 - 4)(-4 - 1) = 8 \times -3 = -24$). Isso indica uma situação onde a função tende ao infinito, pois estamos dividindo um número por um valor que se aproxima de zero.

Isso significa que, conforme $x$ se aproxima de $-4$, a expressão tende a um valor negativo infinitamente grande, devido ao comportamento dos termos fatorados e como eles influenciam a função na vizinhança de $x = -4$.

$$\lim_{x \to -4} \frac{x^2 - 5x + 4}{x^2 + 3x - 4} = -\infty$$

Esta conclusão reflete o comportamento da função próxima a $x = -4$, onde o denominador se aproxima de zero de forma que a função tende a um valor negativamente infinito.

Neste caso, obrigatoriamente, precisamos observar este limite nos aproximando pela esquerda e pela direita. Como pode ser visto no código a seguir

```python
from sympy import symbols, limit, diff, sin, cos, ln, exp, sqrt, Rational
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = symbols('x')
f = (x**2 -5*x+4)/(x**2+3*x - 4)

ponto_desejado = -4

def criar_pontos_aproximacao(valor_limite, distancia=[0.1, 0.01, 0.001, 0.0001, 0.00001]):
    pontos_esquerda = [valor_limite - d for d in distancia]
    pontos_direita = [valor_limite + d for d in distancia]
    return pontos_esquerda, pontos_direita

# Reaplicando a função para criar os pontos de aproximação
pontos_esquerda, pontos_direita = criar_pontos_aproximacao(ponto_desejado)

# Recalculando os valores de f(x) para os pontos à esquerda e à direita
valores_esquerda = [f.subs(x, ponto).evalf() for ponto in pontos_esquerda]
valores_direita = [f.subs(x, ponto).evalf() for ponto in pontos_direita]

# Preparando a saída em formato de tabela novamente
tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

display(Math(tabela_latex))

```

Observe, no código acima que o valor da função cresce exponencialmente a medida que nos aproximamos de $x=-4$ em direções opostas na função. Sendo assim, a
resposta correta seria:

$$\lim_{x \to (-4)^+} \frac{x^2 - 5x + 4}{x^2 + 3x - 4} = -\infty$$
$$\lim_{x \to (-4)^-} \frac{x^2 - 5x + 4}{x^2 + 3x - 4} = \infty$$

você pode confirmar isso usando o Wolfram Alpha em modo linguagem natural com o texto: limit (x^2 - 5x + 4) / (x^2 + 3x - 4) as x -> -4

Podemos confirmar isso vendo o gráfico desta função.


```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo a função
def funcao(x):
    return (x**2 - 5*x + 4) / (x**2 + 3*x - 4)

# Gerando pontos ao redor de -4, mas evitando a divisão por zero
x_vals = np.linspace(-6, -2, 4000)
y_vals = funcao(x_vals)

# Criando o gráfico
fig, ax = plt.subplots(figsize=(12, 6))

# Plotando a função
ax.plot(x_vals, y_vals, label='$f(x) = \\frac{x^2 - 5x + 4}{x^2 + 3x - 4}$')

# Linha pontilhada em x = -4
ax.axvline(x=-4, color='red', linestyle='--', label='x = -4')

# Definindo limites para visualização
ax.set_ylim(-100, 100)

# Adicionando setas nos eixos
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Adicionando setas
ax.plot((1), (0), ls="", marker=">", ms=10, color="k",
        transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0), (1), ls="", marker="^", ms=10, color="k",
        transform=ax.get_xaxis_transform(), clip_on=False)

# Legendas e títulos
ax.set_title('Comportamento de $f(x)$ próximo a $x = -4$')
ax.legend()

plt.show()

```

- O numerador $x^2 + 3x + 2$ pode ser fatorado como $(x + 2)(x + 1)$.
- O denominador $x^2 - 4$ é uma diferença de quadrados e pode ser fatorado como $(x - 2)(x + 2)$.

$$\lim_{x \to -2} \frac{(x + 2)(x + 1)}{(x - 2)(x + 2)}.$$

Assim, a expressão simplificada é:

$$\lim_{x \to -2} \frac{x + 1}{x - 2}.$$

Portanto, o limite da função original quando $x$ se aproxima de $-2$ é $\frac{1}{4}$.

## Problema 6

$$\lim_{x \to 0} \frac{x^2 - 9}{x^2 - x}$$

Vamos resolver o limite

$$\lim_{x \to 0} \frac{x^2 - 9}{x^2 - x}.$$

Primeiro, aplicamos a substituição direta de $x = 0$ na equação:

$$\frac{0^2 - 9}{0^2 - 0} = \frac{-9}{0}.$$

Ao contrário de uma forma indeterminada como por exemplo $\frac{0}{0}$, aqui temos uma divisão por zero, o que indica uma descontinuidade ou uma assíntota vertical na função.

Dado que o denominador se aproxima de zero mais rapidamente do que o numerador se aproxima de qualquer número finito, o limite pode ser considerado como não existindo no sentido convencional, ou seja, a função tende ao infinito ou a um valor não definido conforme $x$ se aproxima de 0.

Vamos simplificar a expressão para tentar entender melhor o comportamento da função em torno de $x = 0$.

Podemos fatorar o numerador e o denominador da seguinte forma:

- O numerador $x^2 - 9$ é uma diferença de quadrados e pode ser fatorado como $(x - 3)(x + 3)$.
- O denominador $x^2 - x$ pode ser fatorado tirando $x$ como fator comum, resultando em $x(x - 1)$.

Portanto, a expressão do limite se torna:

$$\lim_{x \to 0} \frac{(x - 3)(x + 3)}{x(x - 1)}.$$

Nesta forma fatorada podemos ver que, para $x = 0$, o denominador é zero, reforçando a conclusão de que o limite aponta para uma descontinuidade ou uma assíntota vertical, e não para um valor finito que possa ser alcançado através de simplificação algébrica.

Portanto, o comportamento da função em torno de $x = 0$ é dominado pela assíntota vertical e o limite não pode ser simplificado para um valor finito usando métodos algébricos padrão.

Vamos ver o gráfico desta função próximo ao ponto desejado.

Obtemos uma forma indeterminada $\frac{0}{0}$, o que indica que precisamos simplificar a expressão algebricamente para resolver o limite.

Para simplificar, notamos que tanto o numerador quanto o denominador têm $x$ como fator comum. Podemos fatorar e simplificar a expressão da seguinte forma:

- O numerador $x^3 - 6x^2 + 9x$ pode ser fatorado como $x(x^2 - 6x + 9)$.
- O denominador $3x^2 - 4x$ pode ser fatorado como $x(3x - 4)$.

Após fatorar:

$$\lim_{x \to 0} \frac{x(x^2 - 6x + 9)}{x(3x - 4)}.$$

Cancelando o fator comum $x$ (que é válido, pois $x \neq 0$ no contexto de limites), a expressão se torna:

$$\lim_{x \to 0} \frac{x^2 - 6x + 9}{3x - 4}.$$

Notamos que o numerador é um quadrado perfeito: $x^2 - 6x + 9 = (x - 3)^2$. Então, a expressão simplificada é:

$$\lim_{x \to 0} \frac{(x - 3)^2}{3x - 4}.$$

Agora, aplicamos a substituição direta novamente:

$$\frac{(0 - 3)^2}{3(0) - 4} = \frac{9}{-4} = -\frac{9}{4}.$$

Portanto, o limite da função quando $x$ se aproxima de 0 é $-\frac{9}{4}$.

- O denominador $x^2 + 5x + 6$ pode ser fatorado como $(x + 2)(x + 3)$.

Após a fatoração, a expressão se torna:

$$\lim_{x \to -3} \frac{(x - 1)(x - 3)}{(x + 2)(x + 3)}.$$

Percebe-se que o denominador se torna zero quando $x = -3$, confirmando a presença de uma assíntota vertical ou descontinuidade na função nesse ponto, sem uma indeterminação que possa ser simplificada algebricamente para encontrar um valor de limite finito. $$

$$
